```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(10, 5))

# Primary axis for Live Births (Line Plot)
ax1.plot(moh_711_2024.index, moh_711_2024['MOH 711 Live birth'], 
         label='Live Birth', marker='*', color='blue', linewidth=2)
ax1.set_xlabel('Months', fontweight='bold')
ax1.set_ylabel('Live Births', color='blue', fontweight='bold')
ax1.tick_params(axis='y', labelcolor='blue')

# Secondary axis for other birth outcomes (Bar Graph, side-by-side)
ax2 = ax1.twinx()
width = 0.2  # Width of bars
x = np.arange(len(moh_711_2024.index))  # X positions for bars

# Plot bars side by side
ax2.bar(x - width, moh_711_2024['MOH 711 Fresh Still Birth'], 
        width=width, label='Fresh Still Birth', color='red', alpha=0.6)
ax2.bar(x, moh_711_2024['MOH 711 Macerated still Birth'], 
        width=width, label='Macerated Still Birth', color='purple', alpha=0.6)
ax2.bar(x + width, moh_711_2024['MOH 711 Neonatal deaths 0-28 Days'], 
        width=width, label='Neonatal Deaths', color='green', alpha=0.6)

ax2.set_ylabel('Stillbirths & Neonatal Deaths', color='black', fontweight='bold')
ax2.tick_params(axis='y', labelcolor='black')

# Formatting
plt.title('Birth Outcomes 2024', fontweight='bold', fontsize=14)
ax1.set_xticks(x)
ax1.set_xticklabels(moh_711_2024.index, rotation=45, fontweight='bold')

# Legends
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

plt.show()
```
